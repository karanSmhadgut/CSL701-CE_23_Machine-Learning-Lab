Task 9: Fill missing Age values with the median.

```python
# Calculate the median of the Age column
median_age = df["Age"].median()

# Fill missing values with the median
df["Age"] = df["Age"].fillna(median_age)

print("Missing Age values filled successfully.\n")

# Display the first 10 values of the updated Age column
print(df["Age"].head(10))
```

Task 10: Filter surviving passengers (Survived == 1).

```python
survivors = df[df["Survived"] == 1]

print(survivors.head())
print("Number of Survivors:", len(survivors))
```

Task 11: Convert Fare column to a NumPy array.

```python
fare_array = df["Fare"].to_numpy()

print(fare_array)
```

Task 12: Calculate maximum and average Fare using np.max() and np.mean().

```python
max_fare = np.max(fare_array)
avg_fare = np.mean(fare_array)

print("Maximum Fare:", max_fare)
print("Average Fare:", avg_fare)
```

Task 13: Plot histogram of Age.

```python
plt.figure(figsize=(8,5))

plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")

plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

Task 14: Plot bar chart of Survival Count.

```python
survival_count = df["Survived"].value_counts()

plt.figure(figsize=(6,5))

plt.bar(["Not Survived", "Survived"], survival_count,
        color=["red", "green"])

plt.title("Survival Count")
plt.xlabel("Passenger Status")
plt.ylabel("Count")

plt.show()
```

Task 15: Plot scatter graph of Age vs Fare.

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Age"], df["Fare"], color="blue", alpha=0.6)

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```
